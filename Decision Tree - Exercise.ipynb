# build a model to predict if person would survive or not in titanic.csv
- Pclass
- Sex
- Age
- Fare

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
source = df[['Pclass','Sex','Age','Fare']]
source
```

# map Sex column with male value as 1, female as 2

```python
source.Sex = source.Sex.map({'male':1,'female':2})
```

```python
source
```

# Fill Empty age field with mean

```python
source.Age = source.Age.fillna(source.Age.mean())
```

```python
source
```

```python
target = df['Survived']
target
```

# Split the training dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(source,target,train_size=0.8)
```

```python
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

# Decision Tree algorithm

```python
from sklearn import tree
DT = tree.DecisionTreeClassifier()
```

# Fit the model

```python
DT.fit(source,target)
```

```python
X_test
```

# Predict the model for facebook, business , bachelor degree  

```python
DT.predict([[1,1,60.000000,26.5500]])
```

# # Predict the model  

```python
DT.predict(X_test)
```

# Accuracy of the model

```python
#score is 1 bcos the data is less and we have used the same data for training and testing
DT.score(source,target)
```
